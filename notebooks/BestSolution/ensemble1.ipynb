```python
import numpy as np

# 1. Load the two prediction files
# Replace these paths with the actual paths to your files
my_preds = np.load('/kaggle/input/datasets/brijeshwarreddys/previousoutputfiles/preds_bri.npy') 
friend_preds = np.load('/kaggle/input/datasets/brijeshwarreddys/previousoutputfiles/preds_chinki.npy')

# 2. Check shapes - Must be (218, 140, 124, 16)
print(f"My shape: {my_preds.shape}")
print(f"Friend shape: {friend_preds.shape}")

# 3. Weighted Average
# 0.6 weight to PredRNN++ (for episodes), 0.4 to U-Net (for spatial structure)
ensemble_preds = (0.6 * my_preds) + (0.4 * friend_preds)

# 4. Final Safety Clip
# Ensure no floating point math accidentally created negative values
ensemble_preds = np.clip(ensemble_preds, 0, None)

# 5. Save the final submission
np.save('/kaggle/working/preds.npy', ensemble_preds.astype(np.float32))

print("Ensemble Complete! 'preds.npy' is ready for the Top 10 push.")
```
